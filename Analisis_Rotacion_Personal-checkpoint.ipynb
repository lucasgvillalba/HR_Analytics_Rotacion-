```python
# 📦 Librerías necesarias
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from imblearn.over_sampling import SMOTE
import pandas as pd

# 🧹 Cargar y preparar los datos
data = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
data_encoded = pd.get_dummies(data, drop_first=True)

# ✅ Confirmar nombre correcto de la variable objetivo
print(data_encoded.columns)

# 🎯 Variables independientes y dependiente
X = data_encoded.drop('Attrition_Yes', axis=1)
y = data_encoded['Attrition_Yes']

# ✂️ División en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 🧪 Aplicar SMOTE
sm = SMOTE(random_state=42)
X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)

# 🌲 Entrenar modelo Random Forest
model = RandomForestClassifier(random_state=42)
model.fit(X_train_sm, y_train_sm)

# 🔍 Predicciones
y_pred = model.predict(X_test)

# 📊 Resultados
print("Matriz de Confusión:\n", confusion_matrix(y_test, y_pred))
print("\nReporte de Clasificación:\n", classification_report(y_test, y_pred))
print("\nPrecisión del modelo:\n", accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier

# Definir el modelo base
xgb_clf = XGBClassifier(eval_metric='logloss', random_state=42)

# Definir la grilla de parámetros para buscar
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1],
    'subsample': [0.8, 1.0]
}

# Ejecutar búsqueda con validación cruzada
grid_search = GridSearchCV(estimator=xgb_clf, param_grid=param_grid, 
                           cv=3, scoring='accuracy', n_jobs=-1, verbose=1)

grid_search.fit(X_train, y_train)

# Mostrar los mejores parámetros encontrados
print("Mejores parámetros encontrados:")
print(grid_search.best_params_)

# Usar el mejor modelo para predecir
best_xgb = grid_search.best_estimator_
y_pred_xgb_best = best_xgb.predict(X_test)

# Matriz de confusión y métricas
print("Matriz de Confusión:")
print(confusion_matrix(y_test, y_pred_xgb_best))

print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred_xgb_best))

print("Precisión del modelo:")
print(accuracy_score(y_test, y_pred_xgb_best))
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
import numpy as np

# ⚙️ Preprocesamiento y SMOTE
sm = SMOTE(random_state=42)
X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)

# 🔄 Función para suavizar curvas ROC
def suavizar_curva(fpr, tpr):
    unique_fpr, indices = np.unique(fpr, return_index=True)  # Eliminar duplicados
    unique_tpr = tpr[indices]
    f_interpolada = interp1d(unique_fpr, unique_tpr, kind='quadratic')
    fpr_suave = np.linspace(0, 1, 100)
    tpr_suave = f_interpolada(fpr_suave)
    return fpr_suave, tpr_suave

# 📈 Modelos y predicciones
# 1. Logistic Regression
log_model = LogisticRegression(max_iter=10000)  # Aumentar iteraciones
log_model.fit(X_train, y_train)
y_score_log = log_model.predict_proba(X_test)[:, 1]
fpr_log, tpr_log, _ = roc_curve(y_test, y_score_log)
roc_auc_log = auc(fpr_log, tpr_log)
fpr_log_suave, tpr_log_suave = suavizar_curva(fpr_log, tpr_log)

# 2. Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
y_score_rf = rf_model.predict_proba(X_test)[:, 1]
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_score_rf)
roc_auc_rf = auc(fpr_rf, tpr_rf)
fpr_rf_suave, tpr_rf_suave = suavizar_curva(fpr_rf, tpr_rf)

# 3. XGBoost
xgb_model = XGBClassifier(learning_rate=0.1, max_depth=5, n_estimators=50, subsample=0.8, eval_metric='logloss')
xgb_model.fit(X_train, y_train)
y_score_xgb = xgb_model.predict_proba(X_test)[:, 1]
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_score_xgb)
roc_auc_xgb = auc(fpr_xgb, tpr_xgb)
fpr_xgb_suave, tpr_xgb_suave = suavizar_curva(fpr_xgb, tpr_xgb)

# 4. Random Forest + SMOTE
rf_smote_model = RandomForestClassifier(random_state=42)
rf_smote_model.fit(X_train_sm, y_train_sm)
y_score_rf_smote = rf_smote_model.predict_proba(X_test)[:, 1]
fpr_rf_smote, tpr_rf_smote, _ = roc_curve(y_test, y_score_rf_smote)
roc_auc_rf_smote = auc(fpr_rf_smote, tpr_rf_smote)
fpr_rf_smote_suave, tpr_rf_smote_suave = suavizar_curva(fpr_rf_smote, tpr_rf_smote)

# 🖼️ Gráfico de Curvas ROC
plt.figure(figsize=(10, 7))
plt.plot(fpr_log_suave, tpr_log_suave, label=f'Logistic Regression (AUC = {roc_auc_log:.2f})')
plt.plot(fpr_rf_suave, tpr_rf_suave, label=f'Random Forest (AUC = {roc_auc_rf:.2f})')
plt.plot(fpr_xgb_suave, tpr_xgb_suave, label=f'XGBoost (AUC = {roc_auc_xgb:.2f})')
plt.plot(fpr_rf_smote_suave, tpr_rf_smote_suave, label=f'Random Forest + SMOTE (AUC = {roc_auc_rf_smote:.2f})')
plt.plot([0, 1], [0, 1], 'k--', label='Clasificador Aleatorio')

plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curvas ROC suavizadas y comparadas')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
